### Ví dụ về Đa hình (Polymorphism) với Chó và Mèo

Đa hình trong lập trình hướng đối tượng có nghĩa là khả năng một đối tượng có thể mang nhiều hình thái khác nhau hoặc khả năng các đối tượng khác nhau có thể phản ứng khác nhau với cùng một thông điệp (phương thức). Ở đây, chúng ta sẽ định nghĩa một lớp cơ sở `Animal` với một phương thức `speak()`, và sau đó các lớp con `Dog` và `Cat` sẽ *ghi đè* (override) phương thức này để mỗi loài động vật kêu theo cách riêng của mình.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        # Phương thức speak() mặc định của lớp cơ sở
        raise NotImplementedError("Lớp con phải cài đặt phương thức này")

class Dog(Animal):
    def speak(self):
        # Ghi đè phương thức speak() cho chó
        return f"{self.name} nói: Gâu Gâu!"

class Cat(Animal):
    def speak(self):
        # Ghi đè phương thức speak() cho mèo
        return f"{self.name} nói: Meo Meo!"

# Tạo các đối tượng từ các lớp khác nhau
my_dog = Dog("Buddy")
my_cat = Cat("Whiskers")

# Tạo một danh sách chứa các đối tượng động vật
animals = [my_dog, my_cat]

# Gọi phương thức speak() trên từng đối tượng trong danh sách
# Đây là nơi thể hiện tính đa hình: cùng một lời gọi phương thức 'speak'
# nhưng mỗi đối tượng phản ứng khác nhau tùy thuộc vào loại của nó.
for animal in animals:
    print(animal.speak())
```

Trong ví dụ trên:

*   Chúng ta có một `Animal` lớp cơ sở với một phương thức `speak()` chung.
*   `Dog` và `Cat` là các lớp con thừa kế từ `Animal`.
*   Mỗi lớp con **ghi đè** (overrides) phương thức `speak()` để cung cấp cách triển khai riêng cho tiếng kêu của mình.
*   Khi chúng ta lặp qua danh sách `animals` và gọi `animal.speak()`, Python tự động biết phải gọi phương thức `speak()` của lớp `Dog` hay `Cat` dựa trên loại đối tượng hiện tại. Đây chính là bản chất của **đa hình**.

## Kế thừa và Trừu tượng

### Subtask:
Giới thiệu lớp cơ sở trừu tượng `Animal` với phương thức `speak()` trừu tượng, buộc các lớp con `Dog` và `Cat` phải triển khai phương thức này. Điều này minh họa cả tính kế thừa và trừu tượng.


**Reasoning**:
The subtask requires defining an abstract base class `Animal` with an abstract method `speak()`, and then creating concrete subclasses `Dog` and `Cat` that implement this method. This will demonstrate both inheritance and abstraction. I will use the `abc` module for this.



```python
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def speak(self):
        pass # Abstract method, must be implemented by subclasses

class Dog(Animal):
    def speak(self):
        return f"{self.name} nói: Gâu Gâu!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} nói: Meo Meo!"

# Create objects from different classes
my_dog = Dog("Buddy")
my_cat = Cat("Whiskers")

# Create a list containing animal objects
animals = [my_dog, my_cat]

# Call the speak() method on each object in the list
# This demonstrates polymorphism: the same method call 'speak'
# but each object responds differently depending on its type.
for animal in animals:
    print(animal.speak())
```

## Đóng gói (Encapsulation)

### Subtask:
Thêm thuộc tính được đóng gói `_health` vào lớp `Animal` với giá trị ban đầu là 100. Cung cấp các phương thức công khai `get_health()` để truy cập giá trị sức khỏe và `take_damage(amount)` để giảm sức khỏe. Đảm bảo rằng sức khỏe không thể giảm xuống dưới 0.

#### Instructions
1. Add a private attribute `_health` initialized to 100 in the `__init__` method of the `Animal` class.
2. Implement a public method `get_health()` that returns the current `_health` value.
3. Implement a public method `take_damage(amount)` that reduces `_health` by `amount`. Ensure `_health` does not go below 0.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name):
        self.name = name
        self._health = 100  # Encapsulated attribute initialized to 100

    @abstractmethod
    def speak(self):
        pass

    def get_health(self):
        """Public method to get the current health."""
        return self._health

    def take_damage(self, amount):
        """Public method to reduce health, ensuring it doesn't go below 0."""
        if not isinstance(amount, (int, float)) or amount < 0:
            raise ValueError("Damage amount must be a non-negative number.")
        self._health -= amount
        if self._health < 0:
            self._health = 0
        print(f"{self.name} took {amount} damage. Current health: {self._health}")

class Dog(Animal):
    def speak(self):
        return f"{self.name} nói: Gâu Gâu!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} nói: Meo Meo!"

# Create objects from different classes
my_dog = Dog("Buddy")
my_cat = Cat("Whiskers")

# Demonstrate encapsulation
print(f"Initial health of {my_dog.name}: {my_dog.get_health()}")
my_dog.take_damage(30)
print(f"Health of {my_dog.name} after damage: {my_dog.get_health()}")
my_dog.take_damage(80) # Should not go below 0
print(f"Health of {my_dog.name} after more damage: {my_dog.get_health()}")

print(f"Initial health of {my_cat.name}: {my_cat.get_health()}")
my_cat.take_damage(15)
print(f"Health of {my_cat.name} after damage: {my_cat.get_health()}")

# Demonstrate polymorphism again
animals = [my_dog, my_cat]
for animal in animals:
    print(animal.speak())
```
