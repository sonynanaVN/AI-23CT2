### Ví dụ về Đa hình (Polymorphism) với Chó và Mèo

Đa hình trong lập trình hướng đối tượng có nghĩa là khả năng một đối tượng có thể mang nhiều hình thái khác nhau hoặc khả năng các đối tượng khác nhau có thể phản ứng khác nhau với cùng một thông điệp (phương thức). Ở đây, chúng ta sẽ định nghĩa một lớp cơ sở `Animal` với một phương thức `speak()`, và sau đó các lớp con `Dog` và `Cat` sẽ *ghi đè* (override) phương thức này để mỗi loài động vật kêu theo cách riêng của mình.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        # Phương thức speak() mặc định của lớp cơ sở
        raise NotImplementedError("Lớp con phải cài đặt phương thức này")

class Dog(Animal):
    def speak(self):
        # Ghi đè phương thức speak() cho chó
        return f"{self.name} nói: Gâu Gâu!"

class Cat(Animal):
    def speak(self):
        # Ghi đè phương thức speak() cho mèo
        return f"{self.name} nói: Meo Meo!"

# Tạo các đối tượng từ các lớp khác nhau
my_dog = Dog("Buddy")
my_cat = Cat("Whiskers")

# Tạo một danh sách chứa các đối tượng động vật
animals = [my_dog, my_cat]

# Gọi phương thức speak() trên từng đối tượng trong danh sách
# Đây là nơi thể hiện tính đa hình: cùng một lời gọi phương thức 'speak'
# nhưng mỗi đối tượng phản ứng khác nhau tùy thuộc vào loại của nó.
for animal in animals:
    print(animal.speak())
```

Trong ví dụ trên:

*   Chúng ta có một `Animal` lớp cơ sở với một phương thức `speak()` chung.
*   `Dog` và `Cat` là các lớp con thừa kế từ `Animal`.
*   Mỗi lớp con **ghi đè** (overrides) phương thức `speak()` để cung cấp cách triển khai riêng cho tiếng kêu của mình.
*   Khi chúng ta lặp qua danh sách `animals` và gọi `animal.speak()`, Python tự động biết phải gọi phương thức `speak()` của lớp `Dog` hay `Cat` dựa trên loại đối tượng hiện tại. Đây chính là bản chất của **đa hình**.

## Kế thừa và Trừu tượng

### Subtask:
Giới thiệu lớp cơ sở trừu tượng `Animal` với phương thức `speak()` trừu tượng, buộc các lớp con `Dog` và `Cat` phải triển khai phương thức này. Điều này minh họa cả tính kế thừa và trừu tượng.


**Reasoning**:
The subtask requires defining an abstract base class `Animal` with an abstract method `speak()`, and then creating concrete subclasses `Dog` and `Cat` that implement this method. This will demonstrate both inheritance and abstraction. I will use the `abc` module for this.



```python
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def speak(self):
        pass # Abstract method, must be implemented by subclasses

class Dog(Animal):
    def speak(self):
        return f"{self.name} nói: Gâu Gâu!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} nói: Meo Meo!"

# Create objects from different classes
my_dog = Dog("Buddy")
my_cat = Cat("Whiskers")

# Create a list containing animal objects
animals = [my_dog, my_cat]

# Call the speak() method on each object in the list
# This demonstrates polymorphism: the same method call 'speak'
# but each object responds differently depending on its type.
for animal in animals:
    print(animal.speak())
```

## Đóng gói (Encapsulation)

### Subtask:
Thêm thuộc tính được đóng gói `_health` vào lớp `Animal` với giá trị ban đầu là 100. Cung cấp các phương thức công khai `get_health()` để truy cập giá trị sức khỏe và `take_damage(amount)` để giảm sức khỏe. Đảm bảo rằng sức khỏe không thể giảm xuống dưới 0.

#### Instructions
1. Add a private attribute `_health` initialized to 100 in the `__init__` method of the `Animal` class.
2. Implement a public method `get_health()` that returns the current `_health` value.
3. Implement a public method `take_damage(amount)` that reduces `_health` by `amount`. Ensure `_health` does not go below 0.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name):
        self.name = name
        self._health = 100  # Encapsulated attribute initialized to 100

    @abstractmethod
    def speak(self):
        pass

    def get_health(self):
        """Public method to get the current health."""
        return self._health

    def take_damage(self, amount):
        """Public method to reduce health, ensuring it doesn't go below 0."""
        if not isinstance(amount, (int, float)) or amount < 0:
            raise ValueError("Damage amount must be a non-negative number.")
        self._health -= amount
        if self._health < 0:
            self._health = 0
        print(f"{self.name} took {amount} damage. Current health: {self._health}")

class Dog(Animal):
    def speak(self):
        return f"{self.name} nói: Gâu Gâu!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} nói: Meo Meo!"

# Create objects from different classes
my_dog = Dog("Buddy")
my_cat = Cat("Whiskers")

# Demonstrate encapsulation
print(f"Initial health of {my_dog.name}: {my_dog.get_health()}")
my_dog.take_damage(30)
print(f"Health of {my_dog.name} after damage: {my_dog.get_health()}")
my_dog.take_damage(80) # Should not go below 0
print(f"Health of {my_dog.name} after more damage: {my_dog.get_health()}")

print(f"Initial health of {my_cat.name}: {my_cat.get_health()}")
my_cat.take_damage(15)
print(f"Health of {my_cat.name} after damage: {my_cat.get_health()}")

# Demonstrate polymorphism again
animals = [my_dog, my_cat]
for animal in animals:
    print(animal.speak())
```

```python
#"tìm kiếm chiều sâu"
#biểu diễn kiểu dữ liệu
from collections import defaultdict
data = defaultdict(list)

data['A'] = ['B','C','D']
data['B'] = ['E','F']
data['C'] = ['G','H']
data['D'] = ['I','J']
data['F'] = ['K','L','M']
data['H'] = ['N','O']
data['J'] = ['P','Q']
print("Đồ thị:", data)

#khai báo class
class Node:
  def __init__(self,state,parent,action):
    self.state=state
    self.parent=parent

# định nghĩa hàm tìm kiếm theo chiều sâu (DFS)
# S và G được giả định là các đối tượng Node với thuộc tính .state
def DFS(S, G):
  open_list = [S]
  closed_list = set()
  step = 0

  print("\n--- Bắt đầu quá trình tìm kiếm theo chiều sâu (DFS) ---")
  print(f"Nút bắt đầu: {S.state}, Nút đích: {G.state}")

  while True:
    step += 1
    print(f"\nBước {step}:")
    print(f"  Open list hiện tại: {[node.state for node in open_list]}")
    print(f"  Closed list hiện tại: {closed_list}")

    if not open_list:
      print("  Open list trống. Tìm kiếm thất bại.")
      return False

    O = open_list.pop()
    print(f"  Lấy nút '{O.state}' từ open_list để xét.")

    if O.state in closed_list:
        print(f"  Nút '{O.state}' đã được xét. Bỏ qua.")
        continue
    closed_list.add(O.state)

    if O.state == G.state:
      print(f"  Đã chạm tới đích '{G.state}'! Tìm kiếm thành công.")
      path = []
      current = O
      while current:
        path.append(current.state)
        current = current.parent
      print("  Đường đi tìm được:", " ".join(path[::-1])) # In đường đi không dùng ->
      print("--- Kết thúc quá trình tìm kiếm ---")
      return True

    print(f"  Nút '{O.state}' không phải đích. Xét các nút con.")
    if O.state in data: # Kiểm tra xem O.state có con trong đồ thị data không
      print(f"  Các nút con của '{O.state}': {data[O.state]}")
      for child_state in reversed(data[O.state]):
        if child_state not in closed_list:
            is_in_open = False
            for node_in_open in open_list:
                if node_in_open.state == child_state:
                    is_in_open = True
                    break
            if not is_in_open:
                child_node = Node(child_state, O, None)
                open_list.append(child_node)
                print(f"    Thêm nút '{child_state}' vào open_list.")
            else:
                print(f"    Nút '{child_state}' đã có trong open_list. Bỏ qua.")
        else:
            print(f"    Nút '{child_state}' đã có trong closed_list. Bỏ qua.")
    else:
      print(f"  Nút '{O.state}' không có nút con nào.")

# Khởi tạo các nút bắt đầu và đích
start_node = Node('A', None, None)
goal_node = Node('P', None, None)

# Chạy DFS
DFS(start_node, goal_node)
```

## Thuật toán Best-First Search (Tìm kiếm tốt nhất đầu tiên)

Thuật toán Best-First Search (BFS) là một thuật toán tìm kiếm thông tin sử dụng hàm heuristic `h(n)` để ước lượng chi phí từ nút hiện tại đến nút đích, sau đó luôn chọn mở rộng nút có giá trị `h(n)` thấp nhất. Nó ưu tiên thăm dò các đường đi có vẻ 'gần' đích hơn.

Chúng ta sẽ sử dụng lại cấu trúc dữ liệu đồ thị `data` và hàm heuristic `h` đã định nghĩa trước đó.

```python
from queue import PriorityQueue as pq

# Sử dụng lại lớp Node và data/h đã định nghĩa
# class Node:
#     def __init__(self, name, g=0, parent=None):
#         self.name = name
#         self.g = g
#         self.h = h.get(name, float('inf'))
#         self.f = self.g + self.h # Trong Best-First, chúng ta chỉ dùng h để sắp xếp
#         self.parent = parent

#     def __lt__(self, other):
#         return self.h < other.h # Sắp xếp theo h-cost cho Best-First Search

#     def __eq__(self, other):
#         return self.name == other.name if isinstance(other, Node) else False

#     def __hash__(self):
#         return hash(self.name)

# Định nghĩa lại lớp Node cho Best-First Search để sắp xếp theo h
class NodeBFS(Node):
    def __lt__(self, other):
        return self.h < other.h # Sắp xếp theo h-cost cho Best-First Search

def BestFirstSearch(S_name='A', G_name='N'):
    S = NodeBFS(S_name)
    G = NodeBFS(G_name)

    Open = pq()
    Closed = set()

    Open.put(S)

    print(f"\n--- Bắt đầu thuật toán Best-First Search từ '{S_name}' đến '{G_name}' ---")
    step_count = 0

    while True:
        step_count += 1
        print(f"\nBước {step_count}:")
        print(f"  Closed list hiện tại: {sorted(list(Closed))}")

        if Open.empty():
            print("  Open list trống. Không tìm thấy đường đi.")
            return False

        O = Open.get()
        print(f"  Lấy nút '{O.name}' (h={O.h}, g={O.g}) từ Open list để xét.")

        if O.name in Closed:
            print(f"  Nút '{O.name}' đã có trong Closed list. Bỏ qua.")
            continue

        Closed.add(O.name)

        if O.name == G.name:
            print(f"  Đã tìm thấy đích '{G.name}'! Tổng chi phí g: {O.g}")
            path = []
            current = O
            while current:
                path.append(current.name)
                current = current.parent
            print("  Đường đi tìm được:", " -> ".join(path[::-1]))
            print("--- Kết thúc thuật toán Best-First Search ---")
            return True

        print(f"  Nút '{O.name}' không phải đích. Mở rộng các nút con:")
        if O.name in data:
            children_info = data[O.name]
            for i in range(0, len(children_info), 2):
                child_name = children_info[i]
                edge_cost = children_info[i+1]

                tam = NodeBFS(child_name, O.g + edge_cost, O) # Cập nhật g-cost đúng
                print(f"    Xét nút con '{tam.name}' (h={tam.h}, g={tam.g}).")

                if tam.name not in Closed:
                    Open.put(tam)
                    print(f"    Thêm nút '{tam.name}' vào Open list.")
                else:
                    print(f"    Nút '{tam.name}' đã có trong Closed list. Bỏ qua.")
        else:
            print(f"  Nút '{O.name}' không có nút con nào trong đồ thị.")

    print("--- Thuật toán Best-First Search kết thúc (không tìm thấy đường đi) ---")
    return False

# Chạy thuật toán Best-First Search
BestFirstSearch('A', 'N')
```

I will also ensure the `Node` class and data structures are correctly defined for the algorithm.



```python
from collections import defaultdict
from queue import PriorityQueue as pq

# Biểu diễn dữ liệu đồ thị và chi phí cạnh
data = defaultdict(list)
data['A']=['B',2,'C',1,'D',3]
data['B']=['E',5,'F',4]
data['C']=['G',6,'H',3]
data['D']=['I',2,'J',4]
data['F']=['K',2,'L',1,'M',4]
data['H']=['N',2,'O',4]

# Heuristic h(n) - chi phí ước lượng từ nút n đến đích
h = {
'A':6,
'B':5,
'C':4,
'D':5,
'E':7,
'F':4,
'G':6,
'H':2,
'I':7,
'J':6,
'K':3,
'L':4,
'M':5,
'N':0,
'O':4
}

# Khai báo lớp Node để lưu trữ thông tin trạng thái
class Node:
    def __init__(self, name, g=0, parent=None):
        self.name = name
        self.g = g  # Chi phí từ nút bắt đầu đến nút hiện tại
        self.h = h.get(name, float('inf')) # Chi phí ước lượng từ nút hiện tại đến đích
        self.f = self.g + self.h # Tổng chi phí ước lượng (g + h)
        self.parent = parent # Nút cha để tái tạo đường đi

    # Phương thức so sánh để PriorityQueue có thể sắp xếp
    def __lt__(self, other):
        return self.f < other.f

    # Phương thức kiểm tra bằng nhau (cần cho việc tìm kiếm trong tập hợp nếu lưu object)
    def __eq__(self, other):
        return self.name == other.name if isinstance(other, Node) else False

    # Phương thức băm (cần cho việc sử dụng trong tập hợp nếu lưu object)
    def __hash__(self):
        return hash(self.name)

# Hàm tìm kiếm A*
def Astar(S_name='A', G_name='N'):
    # Khởi tạo nút bắt đầu và nút đích
    S = Node(S_name)
    G = Node(G_name)

    # Open list: PriorityQueue lưu trữ các nút cần duyệt, sắp xếp theo f-cost
    Open = pq()
    # Closed list: set lưu trữ tên các nút đã được duyệt
    Closed = set()

    # Bước 1: Cho đỉnh xuất phát vào Open list
    Open.put(S)

    print(f"\n--- Bắt đầu thuật toán A* từ '{S_name}' đến '{G_name}' ---")
    step_count = 0

    while True:
        step_count += 1
        print(f"\nBước {step_count}:")
        # Hiển thị các nút đã đóng (chỉ tên để dễ đọc)
        print(f"  Closed list hiện tại: {sorted(list(Closed))}")

        # Bước 2: Nếu Open list rỗng thì tìm kiếm thất bại
        if Open.empty():
            print("  Open list trống. Không tìm thấy đường đi.")
            return False

        # Bước 3: Lấy đỉnh có f-cost thấp nhất trong Open list ra và gọi đó là O
        O = Open.get()
        print(f"  Lấy nút '{O.name}' (f={O.f}, g={O.g}, h={O.h}) từ Open list để xét.")

        # Nếu nút đã có trong Closed list (có nghĩa là đã tìm thấy đường đi tốt hơn trước đó)
        if O.name in Closed:
            print(f"  Nút '{O.name}' đã có trong Closed list với đường đi tốt hơn hoặc bằng. Bỏ qua.")
            continue

        # Thêm O vào Closed list
        Closed.add(O.name)

        # Bước 4: Nếu O là đỉnh đích thì tìm kiếm thành công
        if O.name == G.name:
            print(f"  Đã tìm thấy đích '{G.name}'! Tổng chi phí: {O.g}")
            path = []
            current = O
            while current:
                path.append(current.name)
                current = current.parent
            print("  Đường đi tìm được:", " -> ".join(path[::-1]))
            print("--- Kết thúc thuật toán A* ---")
            return True

        print(f"  Nút '{O.name}' không phải đích. Mở rộng các nút con:")
        # Bước 5: Tìm tất cả các đỉnh con của O
        if O.name in data:
            children_info = data[O.name]
            for i in range(0, len(children_info), 2):
                child_name = children_info[i]
                edge_cost = children_info[i+1]

                # Tạo một nút con mới
                # Chi phí g mới = chi phí g của nút cha + chi phí cạnh
                tam = Node(child_name, O.g + edge_cost, O)
                print(f"    Xét nút con '{tam.name}' (g={tam.g}, h={tam.h}, f={tam.f}).")

                # Nếu nút con chưa có trong Closed list
                if tam.name not in Closed:
                    # Thêm nút con vào Open list. PriorityQueue sẽ tự động sắp xếp.
                    # Nếu đã có đường đi đến nút này với f-cost cao hơn, nó sẽ bị bỏ qua sau này
                    # khi đường đi mới, tốt hơn được xử lý trước.
                    Open.put(tam)
                    print(f"    Thêm nút '{tam.name}' vào Open list.")
                else:
                    print(f"    Nút '{tam.name}' đã có trong Closed list. Bỏ qua.")
        else:
            print(f"  Nút '{O.name}' không có nút con nào trong đồ thị.")

    print("--- Thuật toán A* kết thúc (không tìm thấy đường đi) ---")
    return False

# Chạy thuật toán A*
Astar('A', 'N')
```
